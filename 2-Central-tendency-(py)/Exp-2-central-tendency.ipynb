```python
# Question 1 mean, median, mode, variance and std dev only
x1 = [6,7,10,12,13,4,8,12]
mean1 = 0
x = sorted(x1)

def mean(lst):
  mean = sum(lst)/len(x)
  print(mean)
print('MEAN')
mean(x)
print('\n')

def mode(lst):
  for i in lst:
    print(i, 'frequency ->', lst.count(i))

print('MODE')
mode(x)
print('\n')

def median(x):
  if len(x)%2 == 0:
    p1 = x[round(len(x)/2)-1]
    p2 = x[round((len(x)/2))]
    print('median for even number of elements: ',p1,' ',p2)
  else:
    p3 = x[round(len(x)/2)]
    print('median for odd number of elements: ',p3)

print('MEDIAN')
median(x)
print('\n')

def dev(lst):
  t = []
  for i in lst:
    t.append((i-9)**2)
  print('Variance: ',sum(t)/len(t))
  print('Std Deviation: ', (sum(t)/len(t))**0.5)
print('MEAN DEVIATION')
dev(x)
print('\n')
```
